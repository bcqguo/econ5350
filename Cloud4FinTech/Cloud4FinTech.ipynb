# Should FinTech embrace Cloud?

* Cloud service providers
  - Amazon Web Services (AWS)
  - Google Cloud Platform (GCP)
  - Microsoft Azure

## The National Institute of Standards and Technology ( www.nist.gov ) defined “Cloud Computing” as:
### A model for enabling ubiquitous, convenient, on-demand network access to a shared pool of configurable computing resources (e.g., networks, servers, storage, applications, and services) that can be rapidly provisioned and released with minimal management effort or service provider interaction.

The way to interact with cloud computing services:
* Software-as-a-Service mode (Saas), such as Dropbox, Gmail, Skype, Twitter, and YouTube
* Platform as a Service (PaaS) provider, such as Amazon owned and operated server farms/data centers and also provided useful tools
* Infrastructure-as-a-Service (IaaS) provider owned and operated the server farms/data centers, but clients would choose their own tools
